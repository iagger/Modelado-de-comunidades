## 1. Importamos las librerias

```python
from sklearn import datasets
import pandas as pd
from pandas.plotting import scatter_matrix
import numpy as np
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import davies_bouldin_score
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

## 2. Cargamos datos

```python
# Cargamos el .csv 

# Datos demograficos
users = 'Prado_users.csv'
demoData = pd.read_csv(users)
# Demografico
print("DEMOGRAFICOS: \n")
print("(Individuos, Num Variables): ", demoData.shape)
print("Nombres variables: ", demoData.keys(), "\n\n")
```

## 3. Evaluación y limpieza de datos

### Análisis de la variable userId (un registro por Id)

```python
# Comprobamos que tenemos un registro por id
repeated = demoData[demoData['userId'].duplicated() == True]

print("Número de userId repetidos: ", repeated['userId'].count())
```

### Análisis de la variable age (categorias, missing, frecuencia)

```python
# Analizamos cuántas categorias de edad se registran.
categoryAge = demoData['age'].value_counts()
print("Categoria edad y frecuencia:\n")
print(categoryAge, "\n")
print("Ilustramos la informacion con un gráfico:\n")
categoryAge.plot.pie()
```

```python
# Comprobamos si hay existe algun missing dentro de la variable edad
print("Numero de missing en la variable age: ",demoData['age'].isnull().sum())
```

### Análisis de la variable gender (categorias, missing, frecuencia)

```python
# Analizamos cuántas categorias de género se registran.
categoryGender = demoData['gender'].value_counts()
print("Categoria género y frecuencia:\n")
print(categoryGender, "\n")
print("Ilustramos la informacion con un gráfico:\n")
categoryGender.plot.pie()
```

```python
# Comprobamos si hay existe algun missing dentro de la variable género
print("Numero de missing en la variable gender: ",demoData['gender'].isnull().sum())
```
