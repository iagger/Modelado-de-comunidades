## 1. Importamos las librerias

```python
from sklearn import datasets
import pandas as pd
from pandas.plotting import scatter_matrix
import numpy as np
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import davies_bouldin_score
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

## 2. Cargamos datos

```python
# Cargamos el .csv 

# Datos emociones
emotions = 'Prado_emotions.csv'
emotionsData = pd.read_csv(emotions)
# Museo
print("Emociones: \n")
print("(Individuos, Num Variables): ", emotionsData.shape)
print("Nombres variables: ", emotionsData.keys())
```

## 3. Evaluación y limpieza de datos

### Análisis de la variable userId (categorias, missing, frecuencia)

```python
# Analizamos cuántas categorias de userId se registran.
categoryUserId = emotionsData['userId'].value_counts()
print("Categoria y frecuencia:\n")
print(categoryUserId, "\n")
print("Ilustramos la informacion con un gráfico:\n")
categoryUserId.plot.pie()
```

```python
# Comprobamos si hay existe algun missing dentro de la variable userId
print("Numero de missing en la variable userId: ",emotionsData['userId'].isnull().sum())
```
